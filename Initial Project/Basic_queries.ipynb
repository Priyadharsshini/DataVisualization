<b>QUESTION:</b> Write code that computes the percentage of males and females that are fans of Star Wars.

#### PERCENTAGE OF MALES AND FEMALES THAT ARE FANS OF STAR WARS

```python
# Filtering all the fans
filtered_data = df[df['fan'] == "Yes"]

malepercentdf = len(filtered_data[filtered_data['gender'] == "Male"])/len(df[df['gender'] == "Male"])*100
print("Male Percentage is",malepercentdf)
```

```python
femalepercentdf = len(filtered_data[filtered_data['gender'] == "Female"])/len(df[df['gender'] == "Female"])*100
print("Female Percentage is",femalepercentdf)
```
